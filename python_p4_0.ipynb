Сравним с кодом matlab/octave (ячейка с "%%octave" у вас, вероятно, не запустится! это для демонстрации)

```python
%load_ext oct2py.ipython
```

```python
%%octave
x = linspace(-3.,3.,100);
y = sin(x);
plot(x,y)
```

```python
x = linspace(-3.,3.,100)
y = sin(x)
plot(x,y)
grid(True)
```

```python
%%octave
x = linspace(-3.,3.,100);
y = sin(x);
plot(x,y)
grid('on')
```
