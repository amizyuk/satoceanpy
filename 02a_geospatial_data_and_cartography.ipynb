Если хотите закрашивать материковую часть, то нужно сделать следующее

```python
import cartopy.feature as cf
```

```python
LAND = cf.NaturalEarthFeature(category='physical', scale='110m', name='land')
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection= ccrs.Mercator())
ax.coastlines('110m')
ax.add_feature(LAND)
```

По аналогии с разрешением берегового контура если указывать scale 50m или 10m, то предварительно эти shape файлы будут скачиваться...
